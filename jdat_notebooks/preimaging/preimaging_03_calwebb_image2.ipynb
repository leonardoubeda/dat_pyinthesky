# Creating simulated images of the JWST astrometric calibration field in the LMC


In this tutorial we show step by step instructions to simulate images of the JWST LMC astrometric calibration field. The NIRCam images are simulated using the software MIRAGE. The observation is designed in APT. The APT output is used as input of MIRAGE.

This Notebook must be executed from an environment that has the JWST pipeline installed.

```python
from sys import exit
import numpy as np
import os
import json
from astropy.io import fits
import glob


from jwst.pipeline import Detector1Pipeline
from jwst.pipeline import Image2Pipeline
from jwst.pipeline import Image3Pipeline
from jwst.pipeline import calwebb_image2
from astropy.table import Table
```

```python
inputtable = 'values-f444w.txt'
```

```python
t = Table.read(inputtable, format='ascii')
inputyaml = t['file']
namelist = []
outlist = []
print(t)
```

```python
level2=True
if level2 == True:
    for k in range(len(inputyaml)):
        print(inputyaml[k])
        base=os.path.basename(inputyaml[k])
        print(base)
        os.path.splitext(base)
        name = os.path.splitext(base)[0]
        exi = os.path.exists(name+'_cal.fits')
        if exi == True:
           continue
        if exi == False:
           namelist.append(name+'_rate.fits')
           outlist.append(name+'.fits')
    print(namelist)

    files =   namelist 
    outfile = outlist 
        
    k = 0
    for filename in files:
        print('---------------------------------------------------------')
        print(filename)

        #here we assign the distortion file according to the detector
        if 'nrca1' in files[k]:
            wcs_reffile = 'NRCA1_FULL_from_pysiaf_distortion.asdf'
        if 'nrca2' in files[k]:
            wcs_reffile = 'NRCA2_FULL_from_pysiaf_distortion.asdf'
        if 'nrca3' in files[k]:
            wcs_reffile = 'NRCA3_FULL_from_pysiaf_distortion.asdf'
        if 'nrca4' in files[k]:
            wcs_reffile = 'NRCA4_FULL_from_pysiaf_distortion.asdf'
        if 'nrcb1' in files[k]:
            wcs_reffile = 'NRCB1_FULL_from_pysiaf_distortion.asdf'
        if 'nrcb2' in files[k]:
            wcs_reffile = 'NRCB2_FULL_from_pysiaf_distortion.asdf'
        if 'nrcb3' in files[k]:
            wcs_reffile = 'NRCB3_FULL_from_pysiaf_distortion.asdf'
        if 'nrcb4' in files[k]:
            wcs_reffile = 'NRCB4_FULL_from_pysiaf_distortion.asdf'
        if 'nrca5' in files[k]:
            wcs_reffile = 'NRCA5_FULL_from_pysiaf_distortion.asdf'
        if 'nrcb5' in files[k]:
            wcs_reffile = 'NRCB5_FULL_from_pysiaf_distortion.asdf'

        m  =  calwebb_image2.Image2Pipeline(config_file='calwebb_image2.cfg', output_file= outfile[k])
        help(wcs_reffile)
        m.assign_wcs.override_distortion = wcs_reffile
        m.run(files[k])
        k=k+1

```
